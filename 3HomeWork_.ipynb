1.1 Создать новый признак price_per_sq_lot,
который будет содержать среднюю стоимость за
один кв. метр общей площади


```python

df[' price_per_sq_lot'] = (df['price'] / df['sqft_lot']).mean()
df.head(2)
```

1.2 Создать новый признак delta_renovated,
который будет содержать разницу в годах между
годом реновацией дома и годом постройки дома Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = df['yr_renovated'] - df['yr_built']
df.loc[df['yr_renovated'] == 0, 'delta_renovated'] = 0
df.head(5)

        
```

1.3 Создайте признаки года продажи, месяца продажи

```python
df['sale_year'] = pd.to_datetime(df['date']).dt.year
df['sale_month'] = pd.to_datetime(df['date']).dt.month
df.sample(2)
```

1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns = ['date', 'zipcode', 'lat', 'long'], inplace = True)
df.head(2)
```

3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
df.groupby('price').agg({'bedrooms':'mean'}).sort_values('price')
```

3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('condition').agg({'price': ['min','mean','max']})
```

3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
tabl_caunt = df.pivot_table(
    values = 'condition', 
    index = 'id',
    columns = 'waterfront',
    fill_value = 0)
tabl_caunt.sample(5)
```
